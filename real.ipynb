```python
import yfinance as yf
from prefect import task, flow
```

```python
@task
def descargar_datos(ticker):
    return yf.download(ticker)
```

```python
@flow(log_prints = True)
def flujo_desacarga_precios():
    frame_yahoo_finance = descargar_datos("AAPL")
    print(frame_yahoo_finance)
```

```python
flujo_desacarga_precios()
```
